# Data visualization with Pandas

```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.DataFrame(data=np.random.randint(1,100, (100,3)), columns=['C1','C2','C3'])
```

```python
df.head()
```

```python
df['C1'].plot.area()
```

```python
df['C1'].plot.barh()
```

```python
df.plot.barh()
```

```python
df['C3'].plot.line()
```

```python
df.plot.box()
```

```python
df['C2'].plot.kde()
```

```python
df.head()
```

```python
df.plot.scatter(x='C1', y='C2')
```

```python
df['C1'].plot.hist(bins=5)
```

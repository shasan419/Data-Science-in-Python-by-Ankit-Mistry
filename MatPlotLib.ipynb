# Importing Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

# Basic Plotting , Plotting treminology

```python
x = np.linspace(0,10, 100)
y = np.sin(x)
z = np.cos(x)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,1,1])
ax.plot(x,y, c='red', alpha=0.8, marker='^', lw=2.5, label='Sine')
ax.plot(x,z, c='green', alpha=0.8, marker='x', label='Cosine')

ax.set_xlim(0,10)
ax.set_ylim(-1,1)
ax.legend(loc=0)

ax.set_xlabel('Time')
ax.set_ylabel('Amp')
ax.set_title('My First Matplotlib graph in Jupyter notebook')
```

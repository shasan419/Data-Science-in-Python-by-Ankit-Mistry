# Univariate F S

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression    # there r many score_function like chi2
```

```python
x
```

```python
sel = SelectKBest(score_func=f_regression,k =4)
```

```python
x_sel = sel.fit(x,y)
```

```python
x_sel.scores_
```

```python
x_sel.transform(x)
```

# Principle Component Analysis

```python
from sklearn.decomposition import PCA
```

```python
pce =PCA(n_components=3)
```

```python
pc =pce.fit(x,y)
```

```python
pc.components_
```

```python
pc.explained_variance_
```

```python
pc.transform(x)
```

# Variancce Threshold - Remove feature with low variance

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
vt =VarianceThreshold(threshold=200)
```

```python
vtp =vt.fit(x,y)
```

```python
vtp.variances_
```

```python
vtp.transform(x)
```
